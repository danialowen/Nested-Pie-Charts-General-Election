```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Scot order = Labour, Con, Lib Dem, SNP with 2024 first and 2019 second
scot = ([[37., 5., 6., 9.], [1., 6., 4., 48]])

# Wales order = Labour, Con, Lib Dem, Plaid Cymru
wal  = ([[27., 0., 1., 4.], [22., 14., 0., 4.]])

# England order = Labour, Con, Lib Dem, Reform, Ind, Green
eng  = ([[348., 116., 65., 5., 5., 4], [180., 345., 7., 0., 0., 1]])
```

```python
# Set up figure and axes
fig, ax = plt.subplots()

size = 0.3

# Plot Soctland results for 2024
ax.pie(scot[0], radius=1, colors=(['#dc0202', '#027edc', '#ff9900', '#ffd241']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

# Plot Soctland results for 2019
ax.pie(scot[1], radius=1-size, colors=(['#dc0202', '#027edc', '#ff9900', '#ffd241']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

# Set title
ax.set_title('Scotland', y=0.95, pad=0, size=20)
plt.show()
```

```python
# Set up figure and axes
fig, ax = plt.subplots()

size = 0.3

# Plot Wales results for 2024
ax.pie(wal[0], radius=1, colors=(['#dc0202', '#027edc', '#ff9900', '#41ff9c']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

# Plot Wales results for 2019
ax.pie(wal[1], radius=1-size, colors=(['#dc0202', '#027edc', '#ff9900', '#41ff9c']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

# Set title
ax.set_title('Wales', y=0.95, pad=0, size=20)
plt.show()
```

```python
# Set up figure and axes
fig, ax = plt.subplots()

size = 0.3

# Plot England results for 2024
ax.pie(eng[0], radius=1, colors=(['#dc0202', '#027edc', '#ff9900', '#3dd6df', '#f63288', '#10962f']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

# Plot England results for 2019
ax.pie(eng[1], radius=1-size, colors=(['#dc0202', '#027edc', '#ff9900', '#3dd6df', '#f63288', '#10962f']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

# Set title
ax.set_title('England', y=0.95, pad=0, size=20)
plt.show()
```

```python
# Set up figure and axes - 2 rows, 3 columns
fig, ax = plt.subplots(2, 3, figsize=(15,8))

size = 0.3

# Create nupy array to create the blank template 
blank = np.array([[100., 0.], [100., 0.]])

# Plot the blank template
ax[0, 1].pie(blank[0], radius=1, colors=(['#e4e4e4', '#e4e4e4']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

ax[0, 1].pie(blank[1], radius=1-size, colors=(['#e4e4e4', '#e4e4e4']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

# Plot the 2024 and 2019 results for Scotland
ax[1, 0].pie(scot[0], radius=1, colors=(['#dc0202', '#027edc', '#ff9900', '#ffd241']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

ax[1, 0].pie(scot[1], radius=1-size, colors=(['#dc0202', '#027edc', '#ff9900', '#ffd241']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

# Plot the 2024 and 2019 results for Wales
ax[1, 1].pie(wal[0], radius=1, colors=(['#dc0202', '#027edc', '#ff9900', '#41ff9c']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

ax[1, 1].pie(wal[1], radius=1-size, colors=(['#dc0202', '#027edc', '#ff9900', '#41ff9c']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

# Plot the 2024 and 2019 results for England
ax[1, 2].pie(eng[0], radius=1, colors=(['#dc0202', '#027edc', '#ff9900', '#3dd6df', '#f63288', '#10962f']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

ax[1, 2].pie(eng[1], radius=1-size, colors=(['#dc0202', '#027edc', '#ff9900', '#3dd6df', '#f63288', '#10962f']), counterclock=False, startangle=90,
       wedgeprops=dict(width=size, edgecolor='w', linewidth=0.5))

# Set the titles
ax[0, 1].set_title('General Election', y=0.95, pad=0, size=20)
ax[1, 0].set_title('Scotland', y=0.95, pad=0, size=20)
ax[1, 1].set_title('Wales', y=0.95, pad=0, size=20)
ax[1, 2].set_title('England', y=0.95, pad=0, size=20)

# Add text to the template diagram
ax[0, 1].text(0, -0.55, '2019', horizontalalignment='center', size=10,
     verticalalignment='center')
ax[0, 1].text(0, -0.85, '2024', horizontalalignment='center', size=10,
     verticalalignment='center')

# Get rid of the top left and top right plots
ax[0,0].set_axis_off()
ax[0,2].set_axis_off();

#fig.savefig("g_election.png", bbox_inches="tight")
```
